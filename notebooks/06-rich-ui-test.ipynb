# Rich UI Testing - Game Layout

Schnelles Ausprobieren von Rich Layouts fÃ¼r das Game UI

```python
from rich.console import Console
from rich.layout import Layout
from rich.panel import Panel

console = Console()
```

## 1. Einfaches Panel

```python
# Einfaches Panel
console.print(Panel("Hallo Welt"))
```

```python
# Panel mit Titel
console.print(Panel("Inhalt hier", title="ğŸ“ Taverne"))
```

```python
# Panel mit Titel + Subtitle
console.print(Panel(
    "gehen, nehmen, sprechen",
    title="ğŸ’¬ Commands",
    subtitle="> Eingabe hier"
))
```

## 2. Layout - Vertical Split (Ã¼bereinander)

```python
layout = Layout()

# Vertikal teilen
layout.split_column(
    Layout(name="oben"),
    Layout(name="unten")
)

# BefÃ¼llen
layout["oben"].update(Panel("Oben", title="Status"))
layout["unten"].update(Panel("Unten", title="Commands"))

console.print(layout)
```

## 3. Layout - Horizontal Split (nebeneinander)

```python
layout = Layout()

# Horizontal teilen
layout.split_row(
    Layout(name="links", ratio=3, title='TEST'),
    Layout(name="rechts", ratio=1)
)

# BefÃ¼llen
layout["links"].update(Panel("Breiter Bereich (75%)", title="Main"))
layout["rechts"].update(Panel("Schmaler Bereich (25%)", title="Inventory"))

console.print(layout)
```

## 4. Kombiniert - Game Layout

```python
layout = Layout()

# Schritt 1: Horizontal (Main + Inventory)
layout.split_row(
    Layout(name="main", ratio=3),
    Layout(name="inventory", ratio=1)
)

# Schritt 2: Main vertikal teilen
layout["main"].split_column(
    Layout(name="status", size=8),
    Layout(name="commands", size=3)
)

# Schritt 3: BefÃ¼llen
layout["status"].update(
    Panel(
        "ğŸ“ Taverne\n\nEine gemÃ¼tliche Taverne mit rustikalen Tischen.\n\nğŸšª Exits: Marktplatz (N), Keller (S)\nğŸ’ Items: SchlÃ¼ssel, Fackel\nğŸ‘¤ NPCs: Wirt",
        title="Location"
    )
)

layout["inventory"].update(
    Panel(
        "â€¢ Karte\nâ€¢ Kompass\nâ€¢ Beutel",
        title="ğŸ’ Inventar"
    )
)

layout["commands"].update(
    Panel(
        "gehen, nehmen, sprechen, untersuchen, Ã¶ffnen, lesen",
        title="ğŸ’¬ VerfÃ¼gbare Commands",
        subtitle="> Eingabe"
    )
)

console.print(layout)
```

## 6. Mit echten Daten - Beispiel

```python
# Simulierte Daten
location = {
    'name': 'Dunkler Wald',
    'beschreibung': 'Hohe BÃ¤ume versperren das Sonnenlicht.'
}
exits = ['Taverne', 'See', 'HÃ¶hle']
items_here = ['Fackel', 'Seil']
npcs = ['RÃ¤uber']
inventory = ['Schwert', 'Schild', 'Trank']
commands = ['gehen', 'nehmen', 'kÃ¤mpfen', 'fliehen']

# Status zusammenbauen
status_text = f"ğŸ“ {location['name']}\n\n{location['beschreibung']}\n\n"
status_text += f"ğŸšª Exits: {', '.join(exits)}\n"
status_text += f"ğŸ’ Items: {', '.join(items_here) if items_here else 'keine'}\n"
status_text += f"ğŸ‘¤ NPCs: {', '.join(npcs) if npcs else 'keine'}"

layout["status"].update(Panel(status_text, title="Location"))

# Inventory
inv_text = "\n".join(f"â€¢ {item}" for item in inventory)
layout["inventory"].update(Panel(inv_text, title="ğŸ’ Inventar"))

# Commands
cmd_text = ", ".join(commands)
layout["commands"].update(Panel(cmd_text, title="ğŸ’¬ Commands", subtitle="> Eingabe"))

console.print(layout)
```

## 7. Leere Panels - Initial State

```python
# Leeres Layout fÃ¼r Spielstart
layout["status"].update(Panel("Lade Location...", title="ğŸ“ Location"))
layout["inventory"].update(Panel("", title="ğŸ’ Inventar"))
layout["commands"].update(Panel("Lade...", title="ğŸ’¬ Commands"))

console.print(layout)
```
